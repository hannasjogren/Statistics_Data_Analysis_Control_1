## Del 2 ‚Äì Statistisk dataanalys i Python

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st
```

```python
# Background
"""
This analysis aims to understand the factors influencing employee satisfaction and their decision to stay or leave Nytto Maximerarna.
By analyzing employee data, I identify patterns and trends to provide recommendations that can improve employee well-being and reduce turnover.

Focus areas:
- Relationship between satisfaction levels and employee turnover
- Impact of working hours and project workload
- Effects of promotions and workplace accidents
- Importance of salary and department affiliation
"""
```

```python
# Load data Excel-file
df = pd.read_excel("HR_Employee_Data.xlsx", engine="openpyxl")

# Save data to CSV
df.to_csv("HR_Employee_Data.csv", index=False)

# Correct column name
if 'average_montly_hours' in df.columns:
    df.rename(columns={'average_montly_hours': 'average_monthly_hours'}, inplace=True)
```

```python
# Basic exploration
display(df.info())
display(df.describe())
```

```python
# Correlation heatmap
# Explanation: A heatmap is used to visualize the correlation between numerical variables, helping to identify key factors affecting employee turnover.
def plot_correlation_heatmap(df):
    numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
    plt.figure(figsize=(10,6))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation between Variables')
    plt.show()
    
plot_correlation_heatmap(df)
```

```python
# Distribution of employee satisfaction levels
# Explanation: A histogram is chosen to show the distribution of satisfaction levels across employees. This helps identify overall trends and potential dissatisfaction among staff.
plt.figure(figsize=(8,5))
sns.histplot(df['satisfaction_level'], bins=30, kde=True, color='blue')
plt.title('Distribution of Satisfaction Levels')
plt.xlabel('Satisfaction Level')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Satisfaction and employee turnover
# Explanation: A boxplot is used to compare satisfaction levels between employees who stayed and those who left. This helps determine if satisfaction plays a role in turnover.
plt.figure(figsize=(8,5))
sns.boxplot(x='left', y='satisfaction_level', data=df, hue='left', palette='coolwarm', legend=False)
plt.title('Satisfaction Levels of Employees Who Stayed and Left')
plt.xlabel('Left the Company (0 = No, 1 = Yes)')
plt.ylabel('Satisfaction Level')
plt.show()
```

```python
# Salary and employee turnover
# Explanation: A countplot is used to visualize how employee turnover varies by salary level. This helps in understanding if lower wages contribute to higher resignation rates.
plt.figure(figsize=(8,5))
sns.countplot(x=df['salary'], hue=df['left'], palette='coolwarm')
plt.title('Employee Turnover by Salary Level')
plt.xlabel('Salary Level')
plt.ylabel('Number of Employees')
plt.legend(title='Left the Company')
plt.show()
```

```python
# Working hours and employee turnover
# Explanation: A boxplot is used to compare the distribution of working hours among employees who stayed and those who left. This helps reveal if overworking leads to higher turnover.
plt.figure(figsize=(8,5))
sns.boxplot(x='left', y='average_monthly_hours', data=df, hue='left', palette='coolwarm', legend=False)
plt.title('Average Monthly Hours and Employee Turnover')
plt.xlabel('Left the Company (0 = No, 1 = Yes)')
plt.ylabel('Average Monthly Hours')
plt.show()
```

```python
# Number of projects and employee turnover
# Explanation: This boxplot compares the number of projects handled by employees who left vs those who stayed, helping to determine if excessive workload is a factor in turnover.
plt.figure(figsize=(8,5))
sns.boxplot(x='left', y='number_project', data=df, hue='left', palette='coolwarm', legend=False)
plt.title('Number of Projects and Employee Turnover')
plt.xlabel('Left the Company (0 = No, 1 = Yes)')
plt.ylabel('Number of Projects')
plt.show()
```

```python
# Promotion history and employee turnover
# Explanation: A countplot visualizes how promotion history affects turnover. If employees who were not promoted in the last five years tend to leave, this highlights a need for career development policies.
plt.figure(figsize=(8,5))
sns.countplot(x=df['promotion_last_5years'], hue=df['left'], palette='coolwarm')
plt.title('Employee Turnover Based on Promotions in the Last 5 Years')
plt.xlabel('Promoted in the Last 5 Years (0 = No, 1 = Yes)')
plt.ylabel('Number of Employees')
plt.legend(title='Left the Company')
plt.show()
```

```python
# Workplace accidents and employee turnover
# Explanation: This countplot helps to assess whether workplace accidents influence turnover rates.
plt.figure(figsize=(8,5))
sns.countplot(x=df['Work_accident'], hue=df['left'], palette='coolwarm')
plt.title('Employee Turnover Based on Workplace Accidents')
plt.xlabel('Workplace Accident (0 = No, 1 = Yes)')
plt.ylabel('Number of Employees')
plt.legend(title='Left the Company')
plt.show()
```

```python
# Combined visualization of multiple factors affecting employee turnover using bar plots
# Explanation: The purpose of this combined visualization is to compare multiple key factors in one view to better understand their impact on employee turnover.
# By using subplots, it's possible to observe how various factors such as satisfaction levels, working hours, number of projects and salary levels are related to the likelihood of employees leaving the company.
# The use of 'hue' (left) allows us to clearly differentiate between employees who left the company (1) and those who stayed (0), making it easier to analyze the differences.
# Each of these factors is chosen based on its potential impact on turnover:
# 1. Satisfaction Level: A direct relationship with employee happiness and retention.
# 2. Monthly Hours Worked: Overwork or underwork can affect employee well-being and job satisfaction.
# 3. Number of Projects: High workload could lead to burnout and dissatisfaction, leading to turnover.
# 4. Salary Level: Employees with lower salaries might feel undervalued, leading to a higher chance of leaving.
# These visualizations offer a clear and direct way to assess how these factors interact with turnover, helping to form insights to improve employee retention.

plt.figure(figsize=(14, 10))

# 1. Satisfaction level vs Employee turnover
plt.subplot(2, 2, 1)
sns.barplot(x='left', y='satisfaction_level', data=df, hue='left', palette='coolwarm', errorbar=None)
plt.title('Satisfaction Level vs. Employee Turnover')
plt.xlabel('Left the Company (0 = No, 1 = Yes)')
plt.ylabel('Average Satisfaction Level')

# 2. Monthly hours worked vs Employee turnover
plt.subplot(2, 2, 2)
sns.barplot(x='left', y='average_monthly_hours', data=df, hue='left', palette='coolwarm', errorbar=None)
plt.title('Average Monthly Hours vs. Employee Turnover')
plt.xlabel('Left the Company (0 = No, 1 = Yes)')
plt.ylabel('Average Monthly Hours')

# 3. Number of projects vs Employee turnover
plt.subplot(2, 2, 3)
sns.barplot(x='left', y='number_project', data=df, hue='left', palette='coolwarm', errorbar=None)
plt.title('Number of Projects vs. Employee Turnover')
plt.xlabel('Left the Company (0 = No, 1 = Yes)')
plt.ylabel('Number of Projects')

# 4. Salary level vs Employee turnover
plt.subplot(2, 2, 4)
sns.countplot(x='salary', hue='left', data=df, palette='coolwarm')
plt.title('Salary Level vs. Employee Turnover')
plt.xlabel('Salary Level')
plt.ylabel('Number of Employees')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Executive Summary
summary = """
Key Insights:
- Employees with low satisfaction levels are more likely to leave the company.
- Turnover rates are higher among lower salary groups.
- Employees who work either too many or too few hours per month tend to leave more frequently.
- Employees handling more than five projects are more likely to leave, possibly due to workload stress.
- Employees who have not been promoted in five years tend to leave more often.
- Workplace accidents do not show a strong correlation with turnover.

Recommendations:
1. Implement initiatives to improve employee satisfaction, such as feedback mechanisms and work-life balance programs.
2. Review salary distribution to ensure competitive wages.
3. Optimize workload by limiting the number of projects per employee.
4. Evaluate promotion policies to encourage long-term commitment.
5. Continue monitoring and minimizing workplace accidents, even though their direct impact on turnover is minimal.
"""
print(summary)
```
